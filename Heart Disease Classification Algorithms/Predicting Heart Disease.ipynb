```python
df = df.drop(['yr','id'],axis=1)
```

```python
x = df.iloc[:,:-1]  
```

```python
x
```

```python
y = df.iloc[:,11]
```

```python
y
```

```python
#Import sklearn for tarin,test split and also applying for ML algorithm
from sklearn.model_selection import train_test_split
```

```python
#train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=.3, random_state=1)
```

```python
#Applying ML algorithm
#https://analyticsindiamag.com/7-types-classification-algorithms/
```

```python
#import random forst classification algorithm
from sklearn.ensemble import RandomForestClassifier
```

```python
#created object
Rclf = RandomForestClassifier()
```

```python
Rclf.fit(xtrain,ytrain)
```

```python
Rclf.score(xtest,ytest)
```

```python
#import decision tree classification algorithm
from sklearn.tree import DecisionTreeClassifier
```

```python
Clf = DecisionTreeClassifier()
```

```python
Clf.fit(xtrain,ytrain)
```

```python
Clf.score(xtest,ytest)
```

```python
#ogistic Regression
from sklearn.linear_model import LogisticRegression
```

```python
Lclf = LogisticRegression()
```

```python
Lclf.fit(xtrain,ytrain)
```

```python
Lclf.score(xtest,ytest)
```

```python
#Naïve Bayes
from sklearn.naive_bayes import GaussianNB
```

```python
nbclf = GaussianNB()
```

```python
nbclf.fit(xtrain,ytrain)
```

```python
#print(nbclf.predict(xtest))
nbclf.score(xtest,ytest)
```

```python
#Stochastic Gradient Descent
from sklearn.linear_model import SGDClassifier
```

```python
#sgdclf = SGDClassifier(loss='modified_huber', shuffle=True, random_state = 101)
sgdclf = SGDClassifier()
```

```python
sgdclf.fit(xtrain,ytrain)
```

```python
sgdclf.score(xtest,ytest)
```

```python
#K-Nearest Neighbours
from sklearn.neighbors import KNeighborsClassifier
```

```python
knclf = KNeighborsClassifier(n_neighbors = 15)
```

```python
knclf.fit(xtrain,ytrain)
```

```python
knclf.score(xtest,ytest)
```

```python
#Support Vector Machine
from sklearn.svm import SVC
```

```python
svmclf = SVC(kernel='linear', C=0.025, random_state=101)
#svmclf = SVC()
```

```python
svmclf.fit(xtrain,ytrain)
```
